## Loc
<sub>loc: Use para selecionar colunas pelo rótulo(Nome da coluna). Não aceita **intervalos numéricos** a menos que os rótulos sejam números.

**Exemplo:**

-- Você pode selecionar colunas usando os rótulos: --
```python
df.loc[:, ['A', 'C']]

--Isso NÃO funcionará, pois '1:3' não é um rótulo válido:
df.loc[:, 1:3] # Isso resultará em um erro, pois '1:3' é um índice, não um rótulo válido

````python
## Aqui coloco os indices e as colunas que eu quero trazer
df.loc[1:7,['City','PostalCode']]

## Trazendo todas as linhas das colunas City e PostalCode
df.loc[:,['City','PostalCode']]


````

## Iloc
<sub>iloc: Use para selecionar linhas e colunas pelo **índice numérico**. Aceita intervalos de índices.

**Exemplo:**

-- Você pode selecionar colunas usando índices numéricos: --
```python
df.iloc[:, [0, 2]]  # Seleciona as colunas de índice 0 e 2

--Isso NÃO funcionará se você tentar usar rótulos com iloc:
df.iloc[:, ['A', 'C']]  #  Isso resultará em erro , pois isso é o nome da coluna e não o índice

````python
## Aqui estou selecionando os índices nas linhas e as colunas
df.iloc[0:3, 0:3]
## Ou trazendo pela ordem das colunas
df.iloc[:,4:9]
````
